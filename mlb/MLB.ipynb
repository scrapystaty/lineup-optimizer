```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Firefox()
```

```python
url = r"https://www.fangraphs.com/leaders/major-league?startdate=&enddate=&month=0&season1=2023&season=2020&qual=100&type=1"
```

```python
driver.get(url)
```

```python
select = Select(driver.find_element(By.XPATH,r"/html/body/div[1]/div/div/div[2]/div[16]/div[2]/div[3]/div[1]/select"))
```

```python
select.select_by_index(4)
```

```python
src = driver.page_source
parser = BeautifulSoup(src, "lxml")
table = parser.find("div", attrs = {"class": "table-scroll"})
headers = table.findAll('th')
headerlist = [h["data-stat"] for h in headers[1:]]
```

```python
headerlist = [a for a in headerlist if not 'Line Break' in a]
```

```python
rows = table.findAll('tr')[1:]
```

```python
player_stats = [[td.getText().strip() for td in rows[i].findAll('td')[1:]] for i in range(len(rows))]
```

```python
player_stats_blanks_removed = []

for stat in player_stats:
    statArr = []
    for num in stat:
        if not num == '': 
            statArr.append(num)
    player_stats_blanks_removed.append(statArr)
```

```python
stats = pd.DataFrame(player_stats_blanks_removed, columns=headerlist)
```
